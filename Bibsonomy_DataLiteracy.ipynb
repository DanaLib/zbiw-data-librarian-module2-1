```python
import json
import requests
```

```python
url = "https://www.bibsonomy.org/json/search/data%20literacy?items=1000&dublicates=merged"
```

```python
result = requests.get(url)
```

```python
data = result.json()
```

```python
print(json.dumps(data,indent=2))
```

```python
len(data['items'])
```

```python
with open("bibsonomy.xml", "w") as xml:
    print("<?xml version=\"1.0\" encoding=\"UTF-8\"?>", file=xml)
    print("<publications>", file=xml)
    idnumber = 1
    for item in data['items']:
        if item['type'] != "Publication":
            continue
        desired_entries = ['type', 'title', 'year', 'publisher','label', 'url']
        print("\t<publication>", file=xml)
        print(f"\t\t<id>{idnumber}</id>", file=xml)
        for entry in desired_entries:    
            if entry in item:
                print(f"\t\t<{entry}>{item[entry]}</{entry}>", file=xml)
        if "author" in item:
            print("\t\t<authors>", file=xml)
            for author in item["author"]:
                print(f"\t\t\t<author>{author}</author>", file=xml)
                print("\t\t</authors>", file=xml)
            print("\t</publication>", file=xml)
        idnumber = idnumber + 1
    print ("</publications>", file=xml)
```
